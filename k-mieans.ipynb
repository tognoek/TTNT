```python
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.datasets import make_blobs #Ham dung de tao du lieu ngau nhien
from sklearn.cluster import KMeans # Ham da duoc implement cho thuat toan K-means
```

```python
X, _ = make_blobs(n_samples=200, cluster_std=3, n_features=3, random_state=99)
```

```python
X.shape
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
colors = {0 : 'Brown', 1 : 'r', 2 : 'y'}
color = [colors[i] for i in _]
plt.scatter(x1, x2, c = color)
```

```python
kmeans = KMeans(n_clusters=3, random_state=99)
```

```python
# print(X)
kmeans.fit(X)
print(kmeans)
# print(X)
```

```python
kmeans.labels_
```

```python
color = [colors[i] for i in kmeans.labels_]
plt.scatter(X[:, 0], X[:, 1], c = color)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = color, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = "black")
```

```python
dotInput = np.array([[6, -6, 3], [5,-3, 4]])
print(dotInput.reshape(-1, 2))
```

```python
outLable = kmeans.predict(dotInput.reshape(-1, 3))
print(outLable)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = color, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = "black")
colotInput = [colors[i] for i in outLable]
plt.scatter(dotInput[:,0], dotInput[:,1], c=colotInput, s=1000, marker="*")
```
