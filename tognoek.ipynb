```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.array([48, 49, 75, 65, 80, 45]) #Chieu cao
y = np.array([155, 156, 175, 174, 185, 158]) # Can nang
```

```python
print(x)
```

```python
plt.scatter(x, y, 50, c='y', marker='s')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x.reshape(-1, 1), y)
```

```python
reg.coef_[0]
```

```python
reg.intercept_
```

```python
y_pre = reg.intercept_ + reg.coef_[0] * x
y_pre
```

```python
plt.scatter(x, y, c = 'r', marker='o')
plt.plot(x, y_pre)
plt.show()
```

```python
x_test = 65
y_pred_test = x_test * reg.coef_[0] + reg.intercept_
print(y_pred_test)
```

```python
plt.scatter(x, y, c = 'r', marker='o')
plt.plot(x, y_pre)
plt.scatter(x_test, y_pred_test, c='black', marker='s')
plt.show()
```
